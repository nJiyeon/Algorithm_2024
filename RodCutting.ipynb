```python
import time
import matplotlib.pyplot as plt

def CUT_ROD(p,n) :
    if n == 0 :
        return 0
    q = -1000
    for i in range(1,n+1,1) :
        q = max(q,p[i] + CUT_ROD(p,n-i))
    return q

def DP_CUT_ROD(p, n) :
    r = [i for i in range(0,n+1)]
    r[0] = 0
    for j in range(1,n+1,1) :
        q = -1000
        for i in range(1, j+1,1) :
            q = max(q, p[i] + r[j-i])
        r[j] = q
    return r[n]

p = [0, 1, 5, 8, 9, 10, 17, 17, 20, 24, 30,
        40, 41, 45, 48, 49, 50, 57, 60, 64, 70,
        71, 75, 78, 79, 80, 87, 87, 90, 94, 100]

lengths = [5, 10, 15, 20, 25, 30]

cutRodTime = []
dpCutRoadTime = []

for n in lengths:
    start_time = time.time()
    CUT_ROD(p, n)
    end_time = time.time()
    cutRodTime.append(end_time - start_time)

for n in lengths:
    start_time = time.time()
    DP_CUT_ROD(p, n)
    end_time = time.time()
    dpCutRoadTime.append(end_time - start_time)


plt.plot(lengths, cutRodTime, label='CUT_ROD')
plt.plot(lengths, dpCutRoadTime, label='DP_CUT_ROD')
plt.xlabel('Length')
plt.ylabel('Execution Time(sec)')
plt.legend()
plt.grid(True)
plt.show()

```
